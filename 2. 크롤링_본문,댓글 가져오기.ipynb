```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from bs4 import BeautifulSoup
import pandas as pd
import csv
import os
import urllib.parse
import csv

urls = list(df_self['링크'])

# user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"

# 가짜 useragent 만들기
from fake_useragent import UserAgent

# 랜덤으로 생성한 UserAgent 값을 출력한다
ua = UserAgent(verify_ssl=False)
user_agent = ua.random
print(user_agent)

# 유저 에이전트 지정
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-logging"])
options.add_argument('--user-agent=%s' % user_agent)
driver = webdriver.Chrome('./chromedriver.exe', options=options)

#0. 경로 이동
PATH = 'C:/Crawling_work/blind'
os.chdir(PATH)

# 빈 리스트 만들어주기
all_results= []

for i, url in enumerate(urls):
    # 제목 집어넣기
    title = list(df_self['제목'])[i]        

    driver.get(url)
    time.sleep(3)
    
    # 여기서 서버 만료로 에러가 뜰 때 지금까지 한 거 저장하고 끝내기
    # 내용 가져오기
    try:
        content = driver.find_element(By.XPATH, '/html/body/div/div/div/main/section/div/div[1]/div[2]/p').text
    except Exception as e:
        print(e)
        content = None
    
    # 댓글 가져오기
    try:
        # 파싱하기
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')
        comment_tags = soup.select('div.article-comments p.cmt-txt')
        comments = [tag.get_text() for tag in comment_tags]
    except Exception as e:
        print(e)
        comments = []
        
    result = [title, content, comments]

    # CSV 파일 읽기
    with open('C:/Crawling_work/blind/자기계발.csv', 'a', newline='', encoding='utf-8-sig') as csvfile:
        # CSV writer 생성
        csvwriter = csv.writer(csvfile)

        # CSV 파일에 데이터 추가
        csvwriter.writerow(result) 

    if (i+1) % 200 == 0:
        time.sleep(15)
```
